```python
# Import pandas and numpy, DataFrame/Series

import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

#Read the contents of Ateika_Samuel_Movie_Ratings.csv into a DataFrame

Movie_Ratings = (pd.read_csv("D:/IS362_Week _7_Assignment/Ateika_Samuel_Movie_Ratings.csv"))

#Display the contents of Ateika_Samuel_Movie_Ratings.csv DataFrame
Movie_Ratings
```

```python
# Create a new DataFrame
participant_avg = Movie_Ratings.copy()

# Calculate the average rating for each survey particpant
participant_avg["Participant's Average"] = participant_avg[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda"]].mean(axis=1)

#Display the average rating for each survey participant data
participant_avg
```

```python
# Calculate the average rating for each movie
movie_avg = participant_avg[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda","Participant's Average"]].mean(axis=0)

#Append a new row to the DataFrame with the results
movie_avg = participant_avg.append(movie_avg, ignore_index = True)

# Rename the value in Col1, Row5 to Average Movie Rating
movie_avg.loc[5,['Participants']] = ['Average Movie Rating']

#Display average rating for each movie data
movie_avg
```

```python
# 1. Create a new pandas dataframe, with normalized ratings for each user 
# 2. Show the average ratings for each user and each movie.

# Normalization or Min-Max Scaling is used to transform features to be on a similar scale. 
# The new point is calculated as: X_new = (X - X_min)/(X_max - X_min)

# Create a new DataFrame
normalized_rate = Movie_Ratings.copy()
  
# Apply normalization techniques by all columns
column = ['Red 2', 'John Wick: Chapter 2', 'The Grinch', 'Redline', 'The Holiday', 'Kung Fu Panda']
normalized_rate[column] = (normalized_rate[column] - normalized_rate[column].min()) / (normalized_rate[column].max() - normalized_rate[column].min())    

# Calculate the average rating for each survey particpant
normalized_rate["Participant's Average"] = normalized_rate[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda"]].mean(axis=1)

# Display normalized data
normalized_rate
```

```python
# Calculate the average rating for each movie
norm_movie_avg = normalized_rate[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda","Participant's Average"]].mean(axis=0)

# Append a new row to the DataFrame with the results
norm_movie_avg = normalized_rate.append(norm_movie_avg, ignore_index = True)

# Rename the value in Col1, Row5 to Average Movie Rating
norm_movie_avg.loc[5,['Participants']] = ['Average Movie Rating']

# Display normalized average rating for each movie data
norm_movie_avg
```

```python
# 1. Create a new pandas dataframe, with standardized ratings for each user.
# 2. Show the average ratings for each user and each movie

#Standardization or Z-Score Normalization is the transformation of features by subtracting from mean and dividing by standard deviation. 
# This is often called as Z-score.
# The new score is calculated as: X_new = (X - mean)/Std

# copy the data
standardized_rate = Movie_Ratings.copy()
  
# apply standardization techniques to all of the columns
column = ['Red 2', 'John Wick: Chapter 2', 'The Grinch', 'Redline', 'The Holiday', 'Kung Fu Panda']
standardized_rate[column] = (standardized_rate[column] - standardized_rate[column].mean()) / standardized_rate[column].std()    
  

# Calculate the average rating for each survey particpant
standardized_rate["Participant's Average"] = standardized_rate[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda"]].mean(axis=1)    
    
# Display standardized data  
standardized_rate
```

```python
# Calculate the average rating for each movie
std_movie_avg = standardized_rate[["Red 2", "John Wick: Chapter 2", "The Grinch", "Redline","The Holiday","Kung Fu Panda","Participant's Average"]].mean(axis=0)

# Append a new row to the DataFrame with the results
std_movie_avg = standardized_rate.append(std_movie_avg, ignore_index = True)

# Rename the value in Col1, Row5 to Average Movie Rating
std_movie_avg.loc[5,['Participants']] = ['Average Movie Rating']

# Display standardized data
std_movie_avg
```
